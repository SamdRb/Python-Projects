```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
@author: Sam Darabi, GX1300

"""

from IPython.display import display, Image
from IPython.display import SVG
import tensorflow as tf
import pandas as pd
import tensorflow.keras as keras
import numpy as np
from tensorflow.keras import backend
from keras.callbacks import EarlyStopping
from keras.models import Sequential
from keras.layers import Dense , Input , Dropout
from sklearn.model_selection import train_test_split as splt
```

```python
# Reading the Training and Testing Data
training_Data = pd.read_csv("/content/SPC-Training.csv", header = None)
testing_Data = pd.read_csv("/content/SPC-Testing.csv", header = None)
```

```python
#Defining the Training Input (First 11 Columns) and Output (Columns from 10 to 17 will be output) 
X_training = training_Data.iloc[:, 0:10].values
Y_training = training_Data.iloc[:, 10:17].values

X_training = (X_training - X_training.mean())/X_training.std()
```

```python
#Spliting the Training Data for the Sake of Validation
X_train, X_val, Y_train, Y_val = splt(X_training, Y_training, test_size=0.2, random_state=1)
```

```python
#Defining the Testing Input (First 11 Columns) and Output (Columns from 10 to 17 will be output) 
X_testing = testing_Data.iloc[:, 0:10].values
Y_testing = testing_Data.iloc[:, 10:17].values

X_testing= (X_testing - X_testing.mean())/ X_testing.std()
```

```python
kernel_regularizer=tf.keras.regularizers.l2(l=.1)
model = Sequential()

with tf.name_scope("layer_1"):
    model.add(Dense(21, use_bias=True, 
                kernel_initializer=tf.initializers.GlorotUniform() , activation = "relu" , input_shape=(10,)))

model.add(Dropout(.3) )
with tf.name_scope("layer_2"):
    model.add(Dense(14, use_bias=True, 
                kernel_initializer= tf.initializers.GlorotUniform() , activation = "relu"))
model.add(Dropout(.3))
with tf.name_scope("layer_3"):
    model.add(Dense(28 ,  use_bias=True ,
                kernel_initializer= tf.initializers.GlorotUniform() , activation = "relu"))
model.add(Dropout(.2))
with tf.name_scope("layer_4"):
    model.add(Dense(14 ,  use_bias=True, 
                kernel_initializer= tf.initializers.GlorotUniform() , activation = "relu"))  
with tf.name_scope("layer_output"):
    model.add(Dense(7 ,  use_bias=True,  
                kernel_initializer= tf.initializers.GlorotUniform() , activation = "softmax"))
```

```python
model.compile(optimizer = 'adam' ,loss='categorical_crossentropy' , metrics=['accuracy'])
tb_callback = keras.callbacks.TensorBoard(
    log_dir="./graphs", write_graph=True)

train_model = model.fit(X_training,Y_training, batch_size = 32, 
     epochs = 200 , callbacks=[tb_callback] , validation_data=(X_val, Y_val))

test_evaluation = model.evaluate(X_testing, Y_testing)
```
