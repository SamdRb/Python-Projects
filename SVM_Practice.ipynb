```python
import pandas as pd 
from sklearn.datasets import load_digits
```

```python
df0= load_digits()
```

```python
dir(df0)
```

```python
df = pd.DataFrame(df0.data,columns=df0.feature_names)
df.head()
```

```python
df['target'] = df0.target
df.head()
```

```python
df['image_number'] =df.target.apply(lambda x: df0.target_names[x])
df.head()
```

```python
df[df.target==0].head()
```

```python
df[df.target==1].head()
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('pixel_0_3')
plt.ylabel('pixel_0_4')
plt.scatter(df0['pixel_0_3'], df0['pixel_0_4'],color="green",marker='+')
plt.scatter(df1['pixel_0_3'], df1['pixel_0_4'],color="blue",marker='.')
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['target','image_number'], axis='columns')
y = df.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
len(X_train)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
#Tune parameters

# Regularization (C)

model_C = SVC(C=1)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

```python
# Gamma

model_g = SVC(gamma=10)
model_g.fit(X_train, y_train)
model_g.score(X_test, y_test)
```

```python
# Kernel

model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train, y_train)
```

```python
model_linear_kernal.score(X_test, y_test)
```
